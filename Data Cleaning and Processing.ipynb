```python
# Using the melt function to reshape the data
CPI_df = pd.melt(df_CPI, id_vars=['Country'], var_name='Year', value_name='Value')
```

```python
# Creating a dictionary to map current column names to new names
column_rename_mapping = {
    'Value': 'Unemployment Rate',
    'Country Name':'Country',
}

# Using the rename method to rename columns
filtered_df = filtered_df.rename(columns=column_rename_mapping)
display(filtered_df)
```
